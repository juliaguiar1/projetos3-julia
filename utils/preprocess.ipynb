```python
# Função para identificar outliers
def identify_outliers(data):
    numeric_columns = data.select_dtypes(include=['float64', 'int64']).columns
    outlier_columns = []
    for col in numeric_columns:
        Q1 = data[col].quantile(0.25)
        Q3 = data[col].quantile(0.75)
        IQR = Q3 - Q1
        outlier_step = 1.5 * IQR
        if data[(data[col] < Q1 - outlier_step) | (data[col] > Q3 + outlier_step)].shape[0] > 0:
            outlier_columns.append(col)
    return outlier_columns

# Identificar as colunas com outliers
outlier_columns = identify_outliers(data_cleaned)

# Exibir as colunas que contêm outliers
if outlier_columns:
    print(f"Outliers identificados nas seguintes colunas: {', '.join(outlier_columns)}")
else:
    print("Não foram identificados outliers nas colunas numéricas.")
```

Como colunas relacionadas à idade e tempo no abrigo são necessárias para a analise de dados deste dataset, serão removidos apenas os outliers relacionados à horário de saída e entrada no abrigo pois são irrelevantes para nosso estudo.

```python
# Função para remover outliers apenas nas colunas especificadas
def remove_outliers(data, columns):
    for col in columns:
        Q1 = data[col].quantile(0.25)
        Q3 = data[col].quantile(0.75)
        IQR = Q3 - Q1
        outlier_step = 1.5 * IQR
        data = data[~((data[col] < (Q1 - outlier_step)) | (data[col] > (Q3 + outlier_step)))]
    return data

# Colunas das quais removeremos os outliers
columns_to_remove_outliers = ['outcome_hour', 'intake_hour']

# Remover outliers dessas colunas
data_cleaned = remove_outliers(data, columns_to_remove_outliers)


# Exibir o número de registros restantes após a remoção de outliers
print(f"Número de registros restantes após remoção de outliers: {len(data_cleaned)}")
```

One-Hot Encoding
Utilizamos One-Hot Encoding para colunas categóricas nominais onde não há uma ordem implícita entre os valores. Esta técnica cria uma nova coluna binária para cada valor único da categoria, permitindo que o modelo trate cada valor de maneira igual.

Label Encoding
Utilizamos Label Encoding para colunas categóricas ordinais onde há uma ordem implícita entre os valores. Esta técnica converte os valores categóricos em valores numéricos que refletem a ordem implícita entre eles.

Como a coluna "breed" possui 1170 valores únicos, criaremos uma nova coluna chamada "is_mix_breed" que será usada para identificar se o animal tem uma raça misturada ou pura. Esta adaptação será feita da seguinte forma:

Feature Engineering na coluna "color": 

A coluna color do dataset apresenta valores que podem ser compostos por uma única cor ou por múltiplas cores separadas por uma barra (/). 

Para simplificar a análise e reduzir a quantidade de valores únicos nesta coluna, aplicamos os seguintes passos:

1) Processamento de Cores Múltiplas:

- As cores compostas por duas cores diferentes, como "Black/White", foram substituídas pelo valor "multi", indicando que o animal possui múltiplas cores.
- Cores que repetiam o mesmo valor, como "White/White", foram simplificadas para apenas "White".

2) Agrupamento por Categorias:

As cores serão categorizadas em grupos maiores (lista abaixo). Isso reduzirá significativamente o número de valores únicos na coluna, facilitando a análise.

Black
White
Gray/Blue
Brown/Chocolate
Yellow/Gold/Cream
Red/Orange
Patterned
Other_Colors

3) One-Hot Encoding:

A nova coluna categorizada, color_category, receberá One-Hot Encoding, transformando cada categoria em uma variável binária (0 ou 1). Essa transformação será aplicada tanto no dataset padronizado quanto no dataset normalizado.

```python
# Função para processar a coluna 'color'
def process_color(color):
    if '/' in color:
        colors = color.split('/')
        if colors[0] == colors[1]:
            return colors[0].strip()
        else:
            return 'multi'
    return color.strip()

# Aplicar a função de processamento de cores nos datasets padronizados e normalizados
data_scaled['processed_color'] = data_scaled['color'].apply(process_color)
data_normalized['processed_color'] = data_normalized['color'].apply(process_color)

# Verificar a quantidade de valores únicos após o processamento
unique_colors_scaled = data_scaled['processed_color'].unique()
unique_colors_normalized = data_normalized['processed_color'].unique()
```

Apesar de termos diminuido a quantidade de valores únicos ainda há muitos valores. Então realizaremos a categorização dessas cores.

Categorias:
Black: 'Black', 'Black Smoke', 'Black Brindle', 'Black Tabby'
White: 'White', 'Cream', 'Cream Tabby', 'Silver', 'Silver Tabby', 'Silver Lynx Point'
Gray/Blue: 'Gray', 'Blue', 'Blue Tabby', 'Blue Smoke', 'Gray Tabby', 'Blue Cream', 'Blue Merle', 'Blue Point', 'Blue Tick'
Brown/Chocolate: 'Brown', 'Brown Tabby', 'Brown Brindle', 'Brown Merle', 'Brown Tiger', 'Chocolate', 'Chocolate Point', 'Liver Tick'
Yellow/Gold/Cream: 'Yellow', 'Yellow Brindle', 'Gold', 'Buff', 'Apricot', 'Tan', 'Fawn'
Red/Orange: 'Red', 'Red Merle', 'Red Tick', 'Flame Point', 'Orange', 'Orange Tabby', 'Orange Tiger'
Patterned: 'Tricolor', 'Torbie', 'Tortie', 'Tortie Point', 'Calico', 'Calico Point'
Other_Colors: 'multi', 'Green', 'Pink', 'Agouti', 'Lilac Point', 'Sable', 'Lynx Point', 'Seal Point'

```python
# Função para categorizar cores
def categorize_color(color):
    color = color.strip()
    if color in ['Black', 'Black Smoke', 'Black Brindle', 'Black Tabby']:
        return 'Black'
    elif color in ['White', 'Cream', 'Cream Tabby', 'Silver', 'Silver Tabby', 'Silver Lynx Point']:
        return 'White'
    elif color in ['Gray', 'Blue', 'Blue Tabby', 'Blue Smoke', 'Gray Tabby', 'Blue Cream', 'Blue Merle', 'Blue Point', 'Blue Tick']:
        return 'Gray/Blue'
    elif color in ['Brown', 'Brown Tabby', 'Brown Brindle', 'Brown Merle', 'Brown Tiger', 'Chocolate', 'Chocolate Point', 'Liver Tick']:
        return 'Brown/Chocolate'
    elif color in ['Yellow', 'Yellow Brindle', 'Gold', 'Buff', 'Apricot', 'Tan', 'Fawn']:
        return 'Yellow/Gold/Cream'
    elif color in ['Red', 'Red Merle', 'Red Tick', 'Flame Point', 'Orange', 'Orange Tabby', 'Orange Tiger']:
        return 'Red/Orange'
    elif color in ['Tricolor', 'Torbie', 'Tortie', 'Tortie Point', 'Calico', 'Calico Point']:
        return 'Patterned'
    else:
        return 'Other_Colors'

# Aplicar a categorização de cores nos datasets padronizados e normalizados
data_scaled['color_category'] = data_scaled['processed_color'].apply(categorize_color)
data_normalized['color_category'] = data_normalized['processed_color'].apply(categorize_color)

# Verificar a quantidade de valores únicos após a categorização
unique_color_categories_scaled = data_scaled['color_category'].unique()
unique_color_categories_normalized = data_normalized['color_category'].unique()

print(f"Quantidade de categorias de cores únicas (scaled): {len(unique_color_categories_scaled)}")
print(f"Quantidade de categorias de cores únicas (normalized): {len(unique_color_categories_normalized)}")
```
